```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

# 1. Data Loading

```python
# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
```

```python
X
```

```python
y
```

```python
# Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

# 2. Model Training (Linear Model)

```python
# Select a linear model
model1 = LinearRegression()
```

```python
# Train the model
model1.fit(X, y)
```

# 3. Inference (Linear Model)
### Once we have our model in place we can try to predict the response of new observations

```python
# Make a prediction for Cyprus
X_new = [[37_655.2]]  # Cyprus' GDP per capita in 2020
print(model1.predict(X_new)) # outputs [[6.30165767]]
```

# 4. Model Training (k-Nearest Neighbors)
### We replace the linear regressor model with a k-Nearest Neighbors (in this example, k = 3) regresson model

```python
# Select a 3-Nearest Neighbors regression model
from sklearn.neighbors import KNeighborsRegressor

model2 = KNeighborsRegressor(n_neighbors=3)

# Train the model
model2.fit(X, y)

```

# 5. Model inference (k-Nearest Neighbors)

#### If all went well, your model will make good predictions. If not, you may need to use more attributes (employment rate, health, air pollution, etc.), get more or better-quality training data, or perhaps select a more powerful model (e.g., a polynomial regression model).

```python
# Make a prediction for Cyprus
print(model2.predict(X_new)) # outputs [[6.33333333]]
```
